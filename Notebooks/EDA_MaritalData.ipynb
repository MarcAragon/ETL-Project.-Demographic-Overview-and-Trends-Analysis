```python
incoherentes = df[df['AgeEnd'] < df['AgeStart']]
print("Filas con rangos de edad incoherentes:", len(incoherentes))

incoherentes_years = df[df['YearEnd'] < df['YearStart']]
print("Filas con rangos de aÃ±os incoherentes:", len(incoherentes_years))
```

```python
top_countries = df['Country or area'].value_counts().head(10)
top_countries.plot(kind='barh', figsize=(8,5), color='skyblue')
plt.title(' PaÃ­ses con mÃ¡s registros')
plt.xlabel('Cantidad de registros')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Sex', palette='pastel', order=df['Sex'].value_counts().index)
plt.title('DistribuciÃ³n por sexo', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Sexo')
plt.ylabel('Cantidad de registros')
plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()


plt.figure(figsize=(10,6))
ax = sns.countplot(
    data=df,
    x='MaritalStatus',
    order=df['MaritalStatus'].value_counts().index,
    palette='crest'
)


plt.title('DistribuciÃ³n por estado civil', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Estado civil', fontsize=12)
plt.ylabel('Cantidad de registros', fontsize=12)


plt.xticks(rotation=30, ha='right')


for p in ax.patches:
    ax.annotate(
        f'{int(p.get_height()):,}',
        (p.get_x() + p.get_width() / 2, p.get_height()),
        ha='center', va='bottom',
        fontsize=10, color='black', xytext=(0, 3),
        textcoords='offset points'
    )

plt.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
print(" DistribuciÃ³n de variables numÃ©ricas")

fig, axes = plt.subplots(
    nrows=(len(num_cols) // 3) + 1,
    ncols=3,
    figsize=(16, 12)
)
axes = axes.flatten()

for i, col in enumerate(num_cols):
    sns.histplot(df[col].dropna(), bins=25, kde=True, color='steelblue', ax=axes[i])
    axes[i].set_title(col, fontsize=11, fontweight='bold')
    axes[i].grid(axis='y', linestyle='--', alpha=0.4)
    axes[i].set_xlabel('')
    axes[i].set_ylabel('')

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.suptitle("DistribuciÃ³n de variables numÃ©ricas", fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

```python
df_line = (
    df.groupby(['AgeStart', 'Sex'])['DataValue']
    .mean()
    .reset_index()
    .sort_values('AgeStart')
)

sns.set(style='whitegrid', context='talk')


plt.figure(figsize=(10, 6))
sns.lineplot(
    data=df_line,
    x='AgeStart',
    y='DataValue',
    hue='Sex',
    marker='o',
    linewidth=2.5,
    palette='crest'
)


plt.title('ðŸ“ˆ Promedio de DataValue por grupo de edad y sexo', fontsize=16, fontweight='bold', pad=15)
plt.xlabel('Edad inicial del grupo (aÃ±os)', fontsize=12)
plt.ylabel('Valor promedio del indicador (DataValue)', fontsize=12)
plt.xticks(rotation=0)
plt.grid(True, linestyle='--', alpha=0.4)
plt.legend(title='Sexo', title_fontsize=12, fontsize=11, loc='upper right', frameon=True)

plt.tight_layout()
plt.show()
```
