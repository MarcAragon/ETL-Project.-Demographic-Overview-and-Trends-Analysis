```python
lista_paises = df['Country Name'].tolist()
print(lista_paises)
```

```python
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
year_max = df['Year'].max()
year_min = df['Year'].min()
print(year_max)
print(year_min)
```

```python
print("\n Valores nulos por columna (%):")
nulls = df.isnull().mean() * 100
display(nulls[nulls > 0].sort_values(ascending=False))

plt.figure(figsize=(5,6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Mapa de valores nulos')
plt.show()
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Configurar el tamaño del gráfico y estilo
plt.figure(figsize=(14, 10))
df[num_cols].hist(figsize=(14, 10), bins=20, layout=(len(num_cols)//3 + 1, 3))
plt.suptitle("Distribución de variables numéricas", fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()  # Ajusta los espacios entre gráficos automáticamente
plt.show()
```

```python
df.groupby('Country Name')[[
    'Life expectancy at birth, female (years)',
    'Life expectancy at birth, male (years)',
    'Fertility rate, total (births per woman)',
    'Gini index'
]].mean().sort_values('Life expectancy at birth, female (years)', ascending=False).head(10)

country_life = df.groupby('Country Name')['Life expectancy at birth, female (years)'].mean().sort_values(ascending=False).head(10)
country_life.plot(kind='barh', figsize=(8,6), color='skyblue')
plt.title('Top 10 países con mayor esperanza de vida femenina promedio')
plt.xlabel('Años')
plt.show()
```

```python
cols = [
    'Fertility rate, total (births per woman)',
    'Life expectancy at birth, female (years)',
    'Life expectancy at birth, male (years)',
    'Death rate, crude (per 1,000 people)',
    'Birth rate, crude (per 1,000 people)',
    'Population growth (annual %)'
]

corr = df[cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlación entre indicadores demográficos')
plt.show()
```

```python
df['LifeExp_Gap'] = df['Life expectancy at birth, female (years)'] - df['Life expectancy at birth, male (years)']

gap = df.groupby('Country Name')['LifeExp_Gap'].mean().sort_values(ascending=False).head(10)
gap.plot(kind='bar', color='purple')
plt.title('🧍‍♀️🧍‍♂️ Diferencia promedio de esperanza de vida (F - M)')
plt.ylabel('Años')
plt.show()
```

```python
paises = ['Colombia', 'Chile', 'Argentina', 'Peru', 'Mexico']

df_latam = df[df['Country Name'].isin(paises)]

sns.lineplot(data=df_latam, x='Year', y='Life expectancy at birth, female (years)', hue='Country Name', marker='o')
plt.title('Esperanza de vida femenina en Latinoamérica')
plt.ylabel('Años')
plt.show()
```

```python
paises = ['Colombia', 'Chile', 'Argentina', 'Peru', 'Mexico']

df_latam = df[df['Country Name'].isin(paises)]

sns.lineplot(data=df_latam, x='Year', y='Life expectancy at birth, male (years)', hue='Country Name', marker='o')
plt.title('Esperanza de vida Masculina en Latinoamérica')
plt.ylabel('Años')
plt.show()
```

```python
cols = [
    'Fertility rate, total (births per woman)',
    'Life expectancy at birth, female (years)',
    'Life expectancy at birth, male (years)',
    'Gini index'
]

df_time = df.groupby('Year')[cols].mean()

df_time.plot(figsize=(10,6), linewidth=2)
plt.title('Evolución global promedio de indicadores sociales')
plt.ylabel('Valor promedio')
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```
